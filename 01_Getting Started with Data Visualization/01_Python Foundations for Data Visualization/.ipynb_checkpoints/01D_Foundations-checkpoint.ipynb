# Python Foundations for Data Visualization

## Create objects with data

### Football Teams

```python
groceries = ['Bread', 'Milk', 'Spanish Jamon']
```

### Champions Leagues

```python
prices = [1, 1.5, 3]
```

## Import matplotlib library

> **Mat**hematical **Plot**ting **Lib**rary

Matplotlib is a Python library that contains functions to create Data Visualizations

```python
import matplotlib.pyplot as plt
```

### Scatter plot (points)

```python
plt.scatter(x=groceries, y=prices)
```

### Bar plot

```python
plt.bar(x=groceries, height=prices)
```

### Many others

https://matplotlib.org/stable/plot_types/index

## Customize the plot

### Change the size

```python
plt.scatter(x=groceries, y=prices, s=10)
```

```python
plt.scatter(x=groceries, y=prices, s=200)
```

### Change the color

```python
plt.scatter(x=groceries, y=prices, s=200, c='red')
```

```python
plt.scatter(x=groceries, y=prices, s=200, c='orange')
```

### Change the marker

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x=groceries, y=prices, s=200, c='orange', marker='*')
```

```python
plt.scatter(x=groceries, y=prices, s=200, c='orange', marker='X')
```

### Change the border

```python
plt.scatter(x=groceries, y=prices, s=200, c='orange', marker='X', edgecolors='black')
```
